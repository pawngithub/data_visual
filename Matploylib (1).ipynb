# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Draw a line in diagram from position (0,10) to position (5,150)
x=np.array([1,2])
y=np.array([5,10])

plt.plot(x,y)
plt.show
```

```python
x=np.array([1,5])
y=np.array([10,100])
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,5])
y=np.array([10,100])
plt.plot(x,y,'*')
plt.show()
```

```python
ypoints = np.array([2,6,8,12])

plt.plot(ypoints,'*''g')
plt.show()
```

```python
ypoints = np.array([2,6,8,12])

### default it takes x values is 01,2,3,4...
plt.plot(ypoints, marker='o', color='g', ms=10, mec='r', mfc='k')  ### mec marker edge color, ms= marker size, k means black color
## mfc is marker frame color
plt.show()
```

```python
y= np.array([2,5,9,4,6,8])

plt.plot(y, linestyle = 'dashed', marker='o', mec= 'hotpink', linewidth='2', mfc='k', ms= 20)
plt.show()
```

```python
plt.plot(y,linestyle='dotted', marker='o', mfc='g', mec='y', ms=25)
```

```python
xpoints=np.array([10,20,30])
ypoints=np.array([1,2,6])

plt.plot(xpoints,ypoints, ms=10, marker='o', linestyle='dotted', linewidth=3, mec='k', color='y' )
plt.show()
```

```python
y1=np.array([2,5,9])
y2=np.array([4,9,6])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,125])
y=np.array([240,250,260,270,280,290,300.310])

plt.plot(x,'r',)
plt.plot(y,'k')
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,125])
y=np.array([240,250,260,270,280,290,300.310,315,320])

plt.plot(x, y, color='r', linewidth=2)
plt.xlabel("Average Pluse", fontsize=30,color="red")  # Average Pluse is label name
plt.ylabel("Calorie Burnage", fontsize=10,color="red")
plt.title('Sample', fontsize=50, color="red")

plt.show()
```

```python
### composing grid to the graph

x=np.array([80,85,90,95,100,105,110,115,125])
y=np.array([240,250,260,270,280,290,300.310,315,320])

plt.plot(x, y, color='r', linewidth=2)
plt.xlabel("Average Pluse", fontsize=30,color="red")  # Average Pluse is label name
plt.ylabel("Calorie Burnage", fontsize=10,color="red")
plt.title('Sports watch data', fontsize=25, color="red")

plt.grid(color="green")

plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,125])
y=np.array([240,250,260,270,280,290,300.310,315,320])

plt.plot(x, y, color='r', linewidth=2)
plt.xlabel("Average Pluse", fontsize=30,color="red")  # Average Pluse is label name
plt.ylabel("Calorie Burnage", fontsize=10,color="red")
plt.title('Sports watch data', fontsize=25, color="orange")

plt.grid(color="green", linestyle='dotted')

plt.show()
```

```python
x=np.array(['A','B','C'])
y=np.array([3,4,6])

plt.bar(x,y, color='green')

plt.show()
```

```python
x=np.array(['A','B','C'])
y=np.array([3,4,6])

plt.barh(x,y, color='green')

plt.show()
```

```python

y=np.array([3,4,6])

plt.hist(y)

plt.show()
```

```python
x=np.random.normal(100,5,2500)
plt.hist(x)
plt.show()
```

```python
y=np.array([10,30,50,35])

plt.pie(y)
plt.show()
```

```python
y=np.array([10,30,50,35])

mylabels = ['apple', 'orange', 'bananna', 'cheri']



plt.pie(y, labels=mylabels)
plt.show()
```

```python
y=np.array([10,30,50,35])

mylabels = ['apple', 'orange', 'bananna', 'cheri']

myexplode =[0,0.2,0.4,0.8]



plt.pie(y, labels=mylabels, startangle= 90, explode=myexplode)
plt.show()
```

```python
y=np.array([10,30,50,35])

mylabels = ['apple', 'orange', 'bananna', 'cheri']

mycolors =["red", 'orange', 'yellow', 'blue']

plt.pie(y, labels=mylabels, startangle= 90, colors=mycolors)
plt.show()
```

```python
y=np.array([10,30,50,35])

## mylabels = ['apple', 'orange', 'bananna', 'cheri']

plt.pie(y)
plt.legend(labels=mylabels)
plt.show()
```
