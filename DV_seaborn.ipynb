```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/pawngithub/Datasets/main/titanic-training-data.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
sns.catplot(x='Pclass', y='Age', data=df)

### catagorical plot x is catagorical and y is numerical
```

```python
sns.catplot(x='Pclass', y='Age', jitter =False, data=df)

### catagorical plot x is catagorical and y is numerical
```

```python
sns.catplot(x='Pclass', y='Age', hue='Sex', data=df)
```

```python
sns.catplot(x='Pclass', y='Age', hue='Sex', kind="swarm", data=df)
```

```python
sns.catplot(x='Pclass', y='Age', hue='Sex', kind="box", data=df)
```

```python
sns.catplot(x='Pclass', y='Age', hue='Sex', kind="violin", split = True, data=df)
```

```python
sns.catplot(x='Pclass', y='Age', hue='Sex', kind="bar", data=df)
```

```python
sns.catplot(x='Pclass', y='Age', hue='Sex', kind="point", data=df)
```

```python
sns.relplot(x='Age', y='Fare',color='red', data=df)

### x is  numerical attribute (int or float), y is numerical attribute, hue is catagorical attribute
```

```python
sns.relplot(x='Age', y='Fare', hue='Embarked', color='red', data=df) ###embarked is port at which passenger got into the ship
```
